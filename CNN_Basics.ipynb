## Convolutional Neural Networks

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def drawImg(img,title="Image"):
    plt.imshow(img,cmap='gray')
    plt.style.use("seaborn")
    plt.axis("off")
    plt.title(title+str(img.shape))
    plt.show()
```

```python
img_ = cv2.imread("cody2.jpg")
img_ = cv2.cvtColor(img_,cv2.COLOR_BGR2RGB)
img = cv2.resize(img_,(100,100))
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY,)
drawImg(img,"RGB")
drawImg(img_gray,"Grayscale")
```

```python
# Hardcoded filters
def convolution(img,img_filter):
    W = img.shape[0]
    H = img.shape[1]
    
    F = img_filter.shape[0]
    new_img = np.zeros((W-F+1,H-F+1))
    
    for row in range(W-F+1):
        for col in range(H-F+1):
            for i in range(F):
                for j in range(F):
                    new_img[row][col] += img[row+i][col+j]*img_filter[i][j]
                    
                if(new_img[row][col] > 255):
                    new_img[row][col] = 255
                elif(new_img[row][col] < 0):
                    new_img[row][col] = 0
    
    return new_img #Activation Map
```

```python
blur_filter = np.ones((3,3))/9.0

new_img = convolution(img_gray,blur_filter)
drawImg(img_gray)
drawImg(new_img)
```

```python
edge_filter = np.array(([1,0,-1],[1,0,-1],[1,0,-1]))
output1 = convolution(img_gray,edge_filter)
drawImg(output1)
```
